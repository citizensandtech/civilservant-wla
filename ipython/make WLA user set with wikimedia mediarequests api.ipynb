## Data Format

https://docs.google.com/document/d/12qNrO5jqCXuIzoz5OrMkHgaXyv0wcVjQHkcDDCr2dsk/edit

img.actor: participant wiki id (is that correct, MK?)
participate.2014.count: number of images contributed in 2014 contest
participate.2015.count: number of images contributed in 2015 contest
participate.2016.count: number of images contributed in 2016 contest
participate.2017.count: number of images contributed in 2017 contest
participate.2019.count: number of images contributed in 2019 contest
participate.2014: binary if contributed (1) or not (0) in 2014 contest
participate.2015: binary if contributed (1) or not (0) in 2015 contest
participate.2016: binary if contributed (1) or not (0) in 2016 contest
participate.2017: binary if contributed (1) or not (0) in 2017 contest
participate.2019: binary if contributed (1) or not (0) in 2019 contest
lang: {fr, en, not-set, not-en-fr} user's commons language preference at time of collection
winner: won at least one prize in 2014-2019 (these will be removed from randomization and analysis)
img.views: number of views of all images participant has contributed as part of the contest
has.email: Has email "on" on commons (so that will be notified of talk page message)
